```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("../../data/raw/data.csv")
```

```python
data.head()
```

```python
# Drop Booking_ID , arrival_year , arrival_date
```

```python
candidate_data = data.drop(labels=["Booking_ID","arrival_year","arrival_date"],axis=1)
```

```python
candidate_data.shape
```

```python
encoded_data = pd.get_dummies(candidate_data,dtype=int)
```

```python
encoded_data
```

```python
encoded_data.describe()
```

```python
cols = [col for col in encoded_data.columns if encoded_data.describe().loc["max",col] > 1]
```

```python
scaler = StandardScaler()
```

```python
for col in cols:
    scaler = StandardScaler()
    
    encoded_data[col] = scaler.fit_transform(np.asarray(encoded_data[col]).reshape(-1,1))
```
